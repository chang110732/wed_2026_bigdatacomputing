```python
# (4,5) 구조의 이미지 데이터를 생성 => 상하, 좌우 반전시켜 새로운 이미지 만들기
import numpy as np
# 데이터 생성
img = np.arange(20).reshape((4,-1)) # (4,5)
print(img)

# (1) 좌우반전 (axis = 1)
h_filp = np.flip(img, axis = 1)
print(h_filp)

# (2) 상하 반전 (axis = 0)
v_filp = np.flip(img,axis = 0)
print(v_filp)

# (3) 전체 반전 (상하좌우 반전)
all_filp = np.flip(img)
print(all_filp)
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기

```python
# Ex: 3명의 국어, 영어, 수학 점수에 대한 데이터를 난수로 생성해서 원본 데이터, 과목 점수 평균, 학생 점수의 평균에 대한 세 종류의 그래프 분석
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings # 경고 메시지 숨기기
import os

# 1. 환경 설정 (경고 차단 및 한글 폰트)
warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf' # Fixed to absolute path
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)

# 2. 데이터 준비
np.random.seed(42)
scores = np.random.randint(60,100,12).reshape(3,-1) # (3,4) - Fixed .reshape issue
subjects = ['국어', '영어', '수학', '통계']
students = ['영희', '철수', '신한']

# 리덕션 연산 : 평균
subject_avg = np.mean(scores, axis = 0).round(2) # 과목별 평균
student_avg = np.mean(scores, axis = 1).round(2) # 학생별 평균 - Fixed variable name for clarity

# 3. 복수 개 그래프로 시각화
plt.figure(figsize=(16,6))
plt.suptitle("원본 데이터 및 리덕션 데이터 분", fontproperties=font_prop, fontsize = 20)
# (1) 원본 데이터 열지도
plt.subplot(1,3,1) # 그래프의 위치
im = plt.imshow(scores, cmap = "YlGnBu", aspect= 'auto')
plt.title("원본 성적 분포", fontproperties=font_prop, fontsize = 15)
plt.xlabel('과목명', fontproperties=font_prop)
plt.ylabel('학생명', fontproperties=font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties=font_prop) # Fixed subjects variable name
plt.yticks(np.arange(len(students)), students, fontproperties=font_prop)
cbar = plt.colorbar(im) # Fixed typo cororbar to colorbar
cbar.set_label("점수", fontproperties=font_prop)
#(1-1) 특정 좌표에 점수 텍스트를 표기 : plt.text(x,y,str)
for i in range(3): # 행 (y축)
   for j in range(4): # 열 (x축)
    plt.text(j, i, scores[i, j], ha="center", va="center", color="black", fontsize=12 ) # Fixed . in ha and color
# plt.show()

# (2) 과목별 평균 시각화 : 막대그래프
plt.subplot(1,3,2) # 그래프의 위치
plt.bar(subjects, subject_avg, color = ['red', 'green', 'blue', 'yellow'], edgecolor = 'black') # Fixed typo balck to black
plt.title("과목별 평균 분포", fontproperties=font_prop, fontsize = 15)
plt.xlabel('과목명', fontproperties=font_prop)
plt.ylabel('평균점수', fontproperties=font_prop)
plt.xticks(np.arange(len(subjects)), subjects, fontproperties=font_prop) # Fixed subjects variable name
for i, score in enumerate(subject_avg): # 막대 상단에 평균 점수 표시
   plt.text(i, score, str(score), ha="center", va="bottom", color="black", fontsize=12)

#(3) 학생별 평균 시각화 : 수평막대 그래프
plt.subplot(1,3,3) # 그래프의 위치
plt.barh(students, student_avg, color = 'lightgreen', edgecolor = 'black') # Fixed typo balck to black
plt.title("학생별 평균 분포", fontproperties=font_prop, fontsize = 15)
plt.xlabel('평균점수', fontproperties=font_prop)
plt.ylabel('학생명', fontproperties=font_prop)
plt.yticks(fontproperties=font_prop) # Fixed subjects variable name
for i, score in enumerate(student_avg): # 막대 상단에 평균 점수 표시
   plt.text(i, score, str(score), va="center", color="black", fontsize=12, fontweight = 'bold')

plt.tight_layout()

plt.show()
```

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.column_stack(리스트)


```python
#----------------------
# Ex1 1차원 배열간 결합
#-----------------------
import numpy as np
# 데이터 작성
a = np.array([1,2,3]) # (3,)
b = np.array([4,5,6]) # (3,)

# 세로 쌓기 (수직방향) -> 행 추가해서 2차원 배열
res_stack = np.vstack((a,b))
print(res_stack) # (2,3)

# 가로로 붙이기 (수평방향) -> 열 추가해서 1차원 배열
res_hstack = np.hstack((a,b))
print(res_hstack)
res_concat = np.concatenate((a,b), axis = 0)
print(res_concat)

# 주의 column_stack() -> 2차원 배열
res_col = np.column_stack((a,b))
print(res_col)

#----------------------
# Ex2 형태가 다른 배열간 결합
#-----------------------
arr_1 = np.arange(1,4) # (3,)
arr_2 = np.arange(4,10).reshape(2,3) # (2,3)

# 방법1
# (1) arr_1 -> (1,3) - 열의 개수를 일치
arr_1_2d = arr_1.reshape(-1,3)
print(arr_1_2d)
print()
combined_2d = np.concatenate((arr_1_2d, arr_2), axis = 0)
print(combined_2d)

# 방법 2
v_combined = np.vstack((arr_1, arr_2))
print(v_combined)



```
